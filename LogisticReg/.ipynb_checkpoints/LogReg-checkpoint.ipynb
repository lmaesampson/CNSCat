```python
#Separate into train and test (2/3 and 1/3)
randmalbact = malbact.sample(frac=1)
train = randmalbact.head(124)
test = randmalbact.tail(62)
test.shape
```

# Adapted from https://nbviewer.jupyter.org/gist/ceshine/c9a4308384e744f062f5

```python
y = train['diag']
```

```python
x = train[['CMalaria','Mening','Enceph','glasgtot','sbp']]
```

```python
def tinvlogit(x):
    return t.exp(x) / (1 + t.exp(x))
```

```python
logistic_model = pm.Model()

with logistic_model:
    effects = pm.Normal('effects', mu=0, tau=2. ** -2, shape=(1, x.shape[1]))
    
    # expected parameter
    p = tinvlogit(t.sum(effects * x, axis=1))
    
    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Bernoulli('Y_obs', p=p, observed=y)
```

```python
with logistic_model:
    start = pm.find_MAP()
    step = pm.NUTS(scaling=start)
    trace = pm.sample(5000, step, start)
```

```python
pm.traceplot(trace)
```

```python
new_x = test[['CMalaria','Mening','Enceph','glasgtot','sbp']]
```

```python
def invlogit(x):
    return np.exp(x) / (1 + np.exp(x))
```

```python
eta = np.zeros(new_x.shape[0]) #Generate log odds of bacterial by averaging posterior
for row in trace['effects']:
    row=row[0]
    eta = np.add(eta, np.dot(new_x, row.transpose()))
eta = np.divide(eta, trace['effects'].shape[0])
```

```python
p = invlogit(eta) #probability of bacterial infection
p[:20]
```

```python
prediction = pd.DataFrame({'p_bact':p}) #build df to visualize results
prediction['pred_diag'] = (prediction['p_bact'] > .5) * 1
```

```python
prediction["diag"] = test.reset_index(drop=True)["diag"] # add true diagnosis
prediction["CMalaria"] = test.reset_index(drop=True)["CMalaria"]
prediction["Enceph"] = test.reset_index(drop=True)["Enceph"]
prediction["Mening"] = test.reset_index(drop=True)["Mening"]
prediction["glasgtot"] = test.reset_index(drop=True)["glasgtot"]
prediction["sbp"] = test.reset_index(drop=True)["sbp"]
prediction.head(20)  #observe pred_diag compared to diag to see difference between prediction and actual diagnosis
```
